```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import openpyxl

# 设置支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
rc = {'font.sans-serif': 'SimHei',
      'axes.unicode_minus': False}  # 用来正常显示中文标签
# 设置学术化的图片风格
sns.set_style("whitegrid", rc=rc)  #设置绘图风格
sns.set_palette("hls")  #设置颜色主题
sns.set_context("paper")  #设置绘图元素缩放比例
data = pd.read_csv('data/data.csv')
data['City'] = data['City'].astype(str)
data['Year'] = data['Year'].astype(str)

data.head()
```

```python
# 1. 数据合并,将城市合并成苏北、苏中、苏南,苏北地区包括：徐州、连云港、宿迁、淮安、盐城,苏南地区包括南京、苏州、无锡、常州、镇江苏中地区包括扬州、泰州、南通
data['City'] = data['City'].replace(
    ['徐州市', '连云港市', '宿迁市', '淮安市', '盐城市'], '苏北')
data['City'] = data['City'].replace(
    ['南京市', '苏州市', '无锡市', '常州市', '镇江市'], '苏南')
data['City'] = data['City'].replace(['扬州市', '泰州市', '南通市'], '苏中')
data.head()


```

```python
#绘制折线图
plt.figure(figsize=(10, 6))
sns.pairplot(data, diag_kind='kde', kind='reg', markers='+')
plt.xticks(rotation=45)
plt.show()
```
